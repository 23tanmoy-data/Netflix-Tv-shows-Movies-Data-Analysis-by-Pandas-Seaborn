##### Which year highest no. of tv shows & movies were released by Bar-graph

```python
net1 = net.drop_duplicates(inplace = True)
net1 = net.drop_duplicates()
net1
```

```python
# add datetime to the date column  [ using to_datetime function ]

net1['Date'] = pd.to_datetime(net1['Release_Date'])
net1['Date']
```

```python
# it counts the all occurance of all individual years in date column

net1['Date'].dt.year.value_counts()
```

```python
net1['Date'].dt.year.value_counts().plot(kind= 'bar',color='green')
```

```python
# year wise movie count represnt in bar - graph

net1[net1['Category'] == 'Movie']['Date'].dt.year.value_counts().plot(kind='bar',color='purple')
```

```python
# year wise tv-shows count represnt in bar - graph

net1[net1['Category'] == 'TV Show']['Date'].dt.year.value_counts().plot(kind='bar',color='orange')
```

```python
# no of movies and tv shows represent in visuals

net1[net1['Category']=='TV Show']['Date'].dt.year.value_counts().plot(kind='pie')
```

```python
# no of movies and tv shows represent in visuals

net1[net1['Category']=='Movie']['Date'].dt.year.value_counts().plot(kind='pie')
```

```python
# but if we wanna show this together movies + tvshows  then how we gonna do this and plot year wise 

net1.groupby('Category').Category.count()
```

```python
net1.groupby('Category').Category.count().plot(kind='pie')
```
